```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['**', '***', '###!', '****', '*****', 'NULL']
df = pd.read_csv("ocorrencia_2010_2020.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df.head(10)
```

```python
df.drop(['codigo_ocorrencia1', 'codigo_ocorrencia3', 'codigo_ocorrencia4', 'ocorrencia_latitude', 'ocorrencia_longitude',
         'investigacao_aeronave_liberada', 'investigacao_status', 'divulgacao_relatorio_numero', 
         'divulgacao_relatorio_publicado', 'divulgacao_dia_publicacao', 'total_aeronaves_envolvidas', 
         'ocorrencia_saida_pista','ocorrencia_pais'], axis=1, inplace=True)
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo": pa.Column(pa.Int, required=False),
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo": pa.Column(pa.String, nullable=True),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
```

```python
# ocorrencias com classificação == INCIDENTE_GRAVE e o estato == SP
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
# ocorrencias com classificação == INCIDENTE_GRAVE ou estato == SP
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 | filtro2]
```

```python
# ocorrencias com (classificação == INCIDENTE_GRAVE ou classificação == INCIDENTE) e o estato == SP
filtro1 = (df.ocorrencia_classificacao == 'INCIDENTE GRAVE') | (df.ocorrencia_classificacao == 'INCIDENTE')
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
# ocorrencias com (classificação == INCIDENTE_GRAVE ou classificação == INCIDENTE) e o estato == SP com metodo isin
filtro1 = (df.ocorrencia_classificacao.isin('INCIDENTE GRAVE', == 'INCIDENTE'])
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```
